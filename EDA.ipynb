```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
main_df = pd.read_csv('./data_cleaned/main_df.csv')
```

## Counts

### View counts of vehicles by vehicle class

```python
plt.figure(figsize=(20,15))
sns.countplot(data=main_df, x=main_df['vehicle_class'] )
```

###  View counts of vehicles by manufacturer

```python
plt.figure(figsize=(65,15))
sns.countplot(data=main_df, x=main_df['manufacturer'] )
```

### Price range of vehicles by manufacturer

```python
plt.figure(figsize=(65,15))
ax= sns.boxplot(x=main_df['manufacturer'],y=main_df['price (USD)'])
```

### Top 10 fastest cars on GTA

```python
top_cars = main_df[main_df['vehicle_class']!= 'Planes'].sort_values('top_speed_in_game (km/h)', ascending=False).head(10)
```

```python
top_cars
```

### Top 10 fastest Planes on GTA

```python
top_planes = main_df.sort_values('top_speed_in_game (km/h)', ascending=False).head(10)
top_planes
```

### Top of most expensive vehicles by vehicle class

```python
idx= main_df.groupby(['vehicle_class'])['price (USD)'].idxmax()
```

```python
most_expensive = main_df.loc[idx, [ 'title', 'vehicle_class', 'manufacturer', 'price (USD)']]
```

```python
most_expensive = most_expensive.sort_values(['vehicle_class', 'price (USD)'], ascending=[True, False])
```
